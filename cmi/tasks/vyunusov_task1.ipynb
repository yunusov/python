```python
import sys
import requests
import pandas as pd
from bs4 import BeautifulSoup

url = 'https://wiki.python.org/moin/'
ok_code = 200

def get_text_from_tags(tags):
    result = []
    for tag in tags:
        if tag.text != '':
            result.append(tag.text)
    return result

def get_href_from_tags(tags):
    result = []
    for tag in tags:
        if tag.text != '':
            result.append(tag['href'])
    return result

response = requests.get(url)
status_code = response.status_code

if response.status_code != ok_code:
    print(f'Неверный ответ {status_code}. Прекращение работы программы.')
    sys.exit()

soup = BeautifulSoup(response.text, 'html.parser')

title = soup.title.text
h1 = soup.find_all('h1')
h2 = soup.find_all('h2')
h3 = soup.find_all('h3')

ps = [p.get_text() for p in soup.find_all('p')]
li_list = [li.text for li in soup.find_all('li')]

a_tag = soup.find_all('a')

print('Заголовок:', title)
print('Главные заголовки H1:', get_text_from_tags(h1))
print('Заголовки H2:', get_text_from_tags(h2))
print('Заголовки H3:', get_text_from_tags(h3))
print("Текст страницы: ", ps[0])
print("Списки:", li_list)
print('Текст ссылок:', get_text_from_tags(a_tag))
print('URL ссылок:', get_href_from_tags(a_tag))

print('')
print('Таблицы:')
print('')
tables = soup.find_all('table')
for table in tables:
    rows = []
    for tr in table.find_all('tr'):
        cols = [td.get_text(strip=True) for td in tr.find_all(['td','th'])]
        if cols:
            rows.append(cols)

    df = pd.DataFrame(rows[1:], columns=rows[0])
    display(df)






```
