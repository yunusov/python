```python
def even_odd(*args):
    evens = [x for x in args if x % 2 == 0]
    odds = [x for x in args if x % 2 != 0]
    return evens, odds
print(even_odd(1, 2, 3, 4, 7))  # ([2, 4], [1, 3, 7])
```

```python
def longest_word(*args):
    return max(args, key=len) if args else None
print(longest_word("a", "big", "elephant"))  # "elephant"
```
